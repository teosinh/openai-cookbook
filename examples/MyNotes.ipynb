```python
import fitz  # PyMuPDF
import os

def convert_pdf_to_text(pdf_path):
    # Open the PDF file
    pdf_document = fitz.open(pdf_path)
    text = ""

    # Iterate through each page
    for page_num in range(pdf_document.page_count):
        page = pdf_document.load_page(page_num)
        text += page.get_text()

    return text

# Example usage
pdf_path = "/tmp/officeZip.1722592928715/ECPD/Allied Wire & Cable (Bulk) July-24.pdf"
text = convert_pdf_to_text(pdf_path)
print(text)
```

```python
from dotenv import load_dotenv
%pip install llama-index
%pip install llama-index
from llama_index import LlamaIndex
from llama_index import LlamaIndex

# Sample function to use LlamaIndex for analysis
def analyze_document(text):
    # Initialize the LlamaIndex with your text data
    index = LlamaIndex()
    index.add_documents([text])

    # Perform analysis (e.g., querying information)
    query = "What are the price trends in the document?"
    results = index.query(query)

    return results

# Example usage
results = analyze_document(text)
print(results)
```
